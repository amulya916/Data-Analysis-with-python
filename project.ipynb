Cleaning Data

```python
cols_with_nulls = data.columns[data.isnull().any()].tolist()
cols_with_nulls
```

```python
data = data.dropna()
data = data.reset_index(drop=True)
print("Shape after removing null rows:", data.shape)
```

Analyze the Trend of EV Registrations Over Time

```python
data['Model Year'] = pd.to_numeric(data['Model Year'], errors='coerce')
data['Model Year'] = data['Model Year'].astype(int)
yearly_counts = data['Model Year'].value_counts().sort_index()
print(yearly_counts)
```

```python
cumulative_counts = yearly_counts.cumsum()

plt.figure(figsize=(12, 6))
sns.lineplot(x=cumulative_counts.index, y=cumulative_counts.values, marker='o', color='green')
plt.title("Cumulative EV Registrations Over Time")
plt.xlabel("Model Year")
plt.ylabel("Cumulative Registrations")
plt.grid(True)
plt.tight_layout()
plt.show()
```

Analyze EV Registrations by County in Washington State

```python
county_counts = data['County'].value_counts().sort_values(ascending=False)
county_counts.head(10)
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x=county_counts.head(15).values, y=county_counts.head(15).index, hue=county_counts.head(15), palette="viridis")
plt.title("Top 10 Counties by EV Registrations")
plt.xlabel("Number of EV Registrations")
plt.ylabel("County")
plt.tight_layout()
plt.show()
```

```python
top_counties = data['County'].value_counts().head(6).index.tolist()

filtered_df = data[data['County'].isin(top_counties)]
sns.set(style="whitegrid")

# Create the catplot
g = sns.catplot(
    data=filtered_df,
    x="Electric Vehicle Type",
    kind="count",
    col="County",        # Facet by County
    hue = "Electric Vehicle Type",
    col_wrap=3,          # 3 columns per row
    height=4,
    aspect=1.2,
    palette="hls"
)
g.set_titles("{col_name} County")
g.set_axis_labels("EV Type", "Number of Registrations")
g.fig.subplots_adjust(top=0.9)
g.fig.suptitle("EV Registration Distribution Across Top Counties in Washington State")

plt.show()
```

Distribution of Electric Vehicle Types

```python
# Count the types
type_counts = data['Electric Vehicle Type'].value_counts()

# Plot
plt.figure(figsize=(12, 6))
sns.barplot(x=type_counts.values, y=type_counts.index,hue=type_counts.values, palette='coolwarm')
plt.title("Distribution of Electric Vehicle Types")
plt.xlabel("Number of Vehicles")
plt.ylabel("EV Type")
plt.tight_layout()
plt.show()
```

```python
# Pie chart
plt.figure(figsize=(3,2))
plt.pie(type_counts.values, labels=type_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette("Spectral"))
plt.title("Proportion of EV Types")
plt.axis('equal')  # Make it a circle
plt.show()
```

Analyzing Interrelationships Between Vehicle Attributes Using Correlation Heatmap

```python
numeric_df = data.select_dtypes(include='number')
correlation = numeric_df.corr()
print(correlation['Electric Range'].sort_values(ascending=False))
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.tight_layout()
plt.show()
```

Analyze the distribution of EV model years to understand how old or new most EVs are.

```python
from datetime import datetime
current_year = datetime.now().year

# Calculate vehicle age
data['Vehicle Age'] = current_year - data['Model Year']
data['Vehicle Age']
```

```python
# Set up side-by-side plots
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Histogram
sns.histplot(data['Vehicle Age'], bins=range(0, 21), kde=False, color='skyblue', edgecolor='black', ax=axes[0])
axes[0].set_title("Histogram of EV Age")
axes[0].set_xlabel("Vehicle Age (Years)")
axes[0].set_ylabel("Number of Vehicles")
axes[0].grid(axis='y', linestyle='--', alpha=0.6)

# KDE Plot
sns.kdeplot(data['Vehicle Age'], fill=True, color='orange',bw_adjust=3, ax=axes[1])
axes[1].set_title("KDE of EV Age")
axes[1].set_xlabel("Vehicle Age (Years)")
axes[1].set_ylabel("Density")
axes[1].grid(axis='y', linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```

Estimating Charging Demand Based on Electric Range of EV Type

```python
data['High Charging Demand'] = data['Electric Range'].apply(lambda x: 'Yes' if x < 50 else 'No')

sns.countplot(data=data, x='High Charging Demand', hue='Electric Vehicle Type', palette='Set3')
plt.title("Charging Demand vs EV Type")
plt.xlabel("High Charging Demand (<50 miles range)")
plt.ylabel("Number of Vehicles")
plt.tight_layout()
plt.show()
```
